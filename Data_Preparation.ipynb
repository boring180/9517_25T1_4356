# Load data

The image data will be loaded from Google Drive.

```python
import pickle
import os
import matplotlib.pyplot as plt

data_path = 'drive/MyDrive/9517_25T1_4356/Aerial_Landscapes/'

data = []
label = []

for folder in os.listdir(data_path):
  for f in os.listdir(data_path + folder):
    img = plt.imread(data_path + folder + '/' + f)
    data.append(img)
    label.append(folder)
    print('Load file:' + folder + '/' + f)
```

# Validate the dataset

The dataset will be valid with random choosing 2 images of each class and show if the label is correct.

```python
import random

unique_labels = list(set(label))
validation_data = []
validation_labels = []

for lbl in unique_labels:
    label_indices = [i for i, x in enumerate(label) if x == lbl]
    selected_indices = random.sample(label_indices, min(2, len(label_indices)))  # Select 2 or fewer if available

    for index in selected_indices:
      validation_data.append(data[index])
      validation_labels.append(label[index])
      plt.imshow(data[index])
      plt.title(f"Predicted Label: {label[index]}")
      plt.show()
```

# Train-Test split

10% of the data will be used as test data set, and the rest will be used for training.  

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=0.2, random_state=42, stratify=label)

!mkdir 'drive/MyDrive/9517_25T1_4356/normal_train'
!mkdir 'drive/MyDrive/9517_25T1_4356/normal_test'

for i in range(len(X_train)):
  plt.imsave('drive/MyDrive/9517_25T1_4356/normal_train/' + y_train[i] + '_' + str(i) +'.png', X_train[i])

for i in range(len(X_test)):
  plt.imsave('drive/MyDrive/9517_25T1_4356/normal_test/' + y_test[i] + '_' + str(i) + '.png', X_test[i])
```

# Imbalanced data
For a challenging testing, the training data set will shrink to a imbalanced data set which will foollow a long-tail distribution through different classes.

```python
import numpy as np

labels = np.unique(label)
number_of_labels = len(labels)
print(number_of_labels)

X_train = np.array(X_train)
y_train = np.array(y_train)

!mkdir 'drive/MyDrive/9517_25T1_4356/lt_train'

for i in range(number_of_labels):
  amount_of_img = (y_train == labels[i]).sum() - 40 * i
  print('Processing label: ' + labels[i] + ' With numbe of data ' + str(amount_of_img))
  indices = np.where(y_train == labels[i])[0]
  selected_indices = np.random.choice(indices, amount_of_img, replace=False)
  for j in selected_indices:
    plt.imsave('drive/MyDrive/9517_25T1_4356/lt_train/' + labels[i] + '_' + str(j) +'.png', X_train[j])
```
